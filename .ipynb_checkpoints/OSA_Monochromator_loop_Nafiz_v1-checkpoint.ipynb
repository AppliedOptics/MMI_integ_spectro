## Connect OSA and Mono

```python
import socket
import time
import numpy as np
import csv
import pandas as pd
import matplotlib.pyplot as plt
from scipy.spatial import distance

# import sys
# np.set_printoptions(threshold=sys.maxsize)
```

##  *OSA Function defs* 

```python
def sock_send(sock,msg,receive=False,cond=None):
    sock.send(msg.encode()+b"\r\n")
    if receive:
        if cond != None:
            data = b'0'
            while data.decode("utf-8")[0] != cond:
                sock.send(b":stat:oper:even?\r\n")
                data = sock.recv(1024)

                time.sleep(0.1)
            print(data)
            return data
        else:
            data = sock.recv(1024)
            print(data)
            return data
    return None

def sock_send_arr(sock,msg_list,receive=False):
    ret = []
    for msg in msg_list:
        ret += [sock_send(sock,msg,receive)]
    return ret

def querry_data(sock,msg):
    sock.send(msg.encode()+b"\r\n")  
    strData2=b""
    i=0
    while True:
#         print(f"iter{i}")
        i=i+1
        strData2=sock.recv(1024)
#         print(strData2)
        if strData2.decode("utf-8")[0]=="+":
            break  
    return strData2.decode("utf-8")
```

##  *OSA User Settings* 

```python
#User Inputs
# center_start= 655
# center_end= 693.5 #end point not executed
# center_step=0.5

start_WL=645  #OSA window start
stop_WL=710  #OSA window end
RLevel_nW=400


Resln=0.5
Spn=10
Sampling_interval= 0.05
Average_times=1

sensitivity="high1"  #Options- norm, mid, high1,2,3. generally use high1
smoothing="OFF"
```

```python
# mono_start_wl=655.0
# mono_stop_wl=693.0
# mono_step_wl=0.5
# N=int((mono_stop_wl-mono_start_wl)/mono_step_wl+1) # loop counter

N=120
```

##  *OSA and Mono connect* 

```python
#Connection to OSA
HOST = '192.168.0.1'    # The remote server (OSA)
PORT = 10001              # The same port as used by the server (OSA)
s= socket.socket(socket.AF_INET, socket.SOCK_STREAM)
s.connect((HOST, PORT))    
print("OSA connected")
data = sock_send_arr(s,["open \"anonymous\"", " "],receive=True)

# Connection to Monochromator
HOST_mono='127.0.0.1'  #Monochromator server LabView program (this machines IP)
PORT_mono=27025
s_mono=socket.socket(socket.AF_INET,socket.SOCK_STREAM)
flag1=s_mono.connect((HOST_mono,PORT_mono))
if flag1==None:
    print("Mono connected")
mono_send=b"abc"



# csv field names 
cent1=0
OSA_data={}
fields = ['index','peakX', 'peakY', 'centerWL', 'widthFWHM']
filename = "monochromator peaks_summary.csv"
```

## <span style='color:red'> <span style='background :yellow' >Main Loop (Mono + OSA) </span>

```python
# for cent in np.arange(center_start,center_end,center_step):
for cent in range(N):
    
    msg=s_mono.recv(1024) 
    if msg.decode("utf-8")=="go":
#         print(f"mono received for {cent}")
#         data = sock_send_arr(s,["*RST","CFORM1",f":sens:wav:cent {cent}nm",f":sens:wav:span {Spn}nm",\
#                                 f":sens:band:bwid:res {Resln}nm",f":sens:sens {sensitivity}",f":sens:sett:smo {smoothing}"\
#                                 ":sens:sweep:points:auto off",f":sens:sweep:step {Sampling_interval}nm",\
#                                ":sens:corr:rvel:med air", f":sens:aver:coun {Average_times}",":sens:chop OFF"])
        data = sock_send_arr(s,["*RST","CFORM1",f":sens:wav:star {start_WL}nm",f":sens:wav:stop {stop_WL}nm"])
        data = sock_send_arr(s,[f":sens:band:res {Resln}nm",f":sens:sens {sensitivity}"])
        data = sock_send_arr(s,[f":sens:sett:smo {smoothing}",\
                                ":sens:sweep:points:auto off",f":sens:sweep:step {Sampling_interval}nm",\
                               ":sens:corr:rvel:med air"])
        data = sock_send_arr(s,[f":sens:aver:coun {Average_times}",":sens:chop OFF"])
        data=sock_send_arr(s,[":DISPLAY:TRACE:Y1:SPACING LIN",f":DISPLAY:TRACE:Y1:RLEVEL {RLevel_nW}nW"])
        print(f"mono received for {cent}")
        OSA_data['index']=cent
        data = sock_send_arr(s,[":init:smode 1","*CLS",":init"])
        data = sock_send(s,":stat:oper:even?",receive=True,cond='1')
#         time.sleep(1)
        data = sock_send_arr(s,["CFORM1",":mmem:cdr int"])
        data = sock_send(s,f":mmem:stor:trac tra,csv,\"mono_{cent}\",int",receive=True)
        
        ##calculate peak WL
        data=sock_send(s,":calc:mark:max")
        data_peakx=querry_data(s,":calc:mark:x? 0")        
        OSA_data['peakX']=float(data_peakx)*1e9  #unit nm
        print(f"peak={float(data_peakx)*1e9}")
        data_peaky=querry_data(s,":calc:mark:y? 0")
        OSA_data['peakY']=float(data_peaky)*1e9  #unit nW
        
        #calculate 3dB BW
        data=sock_send_arr(s,[":calc:cat swth"])
        data=sock_send_arr(s,[":calc:par:cat:swth:TH 3.0"])
        data=sock_send(s,":calc")
        str_thresh=querry_data(s,":calc:data?")
        data_thresh=str_thresh.split(',')
        OSA_data['centerWL']=float(data_thresh[0])*1e9
        OSA_data['widthFWHM']=float(data_thresh[1])*1e9
        
        if cent1==0:
            with open(filename, 'w',newline='') as csvfile:
                writer = csv.DictWriter(csvfile, fieldnames = fields) 
                writer.writeheader() 
        cent1+=1        
        with open(filename, 'a',newline='') as csvfile: 
    # creating a csv dict writer object 
            writer = csv.DictWriter(csvfile, fieldnames = fields) 
            writer.writerow(OSA_data) 
        
        s_mono.send(mono_send)
#       print("write successful")
    
flag=s.close()
if flag==None:
    print("Socket Closed Successfully")
    
```

## <span style='color:red'> Option 2: </span> Wavelength match calculation

```python
dir_mono='C:/Users/walrus/Desktop/andor tcp ip_dontDELETE/labview/monoserver/'
dir_NKT='C:/Users/walrus/Desktop/walrus_git/walrus_python_codes/'

dir_mono_outp='C:/Users/walrus/Desktop/spectrometer/Monochromator to MMI_2023_01_19/iHR550/imaging/'
# mono_setpoints=pd.read_csv(dir_mono+'testin.csv',header=None,nrows=77,names=["setpoint"],usecols=[0)
mono_outpoints=pd.read_csv(dir_mono_outp+'testout_combined.csv',header=None,nrows=77,index_col=None,names=["outpoint1","outpoint2","outpoint3",\
                                                                                                 "outpoint4","outpoint5","outpoint6",\
                                                                                                 "outpoint7","outpoint8",\
                                                                                                          "outpoint9","outpoint10","outpoint11"])
mono_list=pd.read_csv(dir_mono+'monochromator peaks_summary.csv',skiprows=0,nrows=78,index_col='index')
NKT_list=pd.read_csv(dir_NKT+'NKTselect_peaks.csv',skiprows=0,nrows=403,index_col='index')

distances=distance.cdist(mono_list['peakX'].to_numpy().reshape(-1,1), NKT_list['peakX'].to_numpy().reshape(-1,1), 'euclidean')
result = np.where(distances == np.amin(distances))
listOfCordinates = list(zip(result[0], result[1]))

# print(mono_outpoints.iloc[3])
mono_outpoints.head()
# mono_setpoints.head()
# print(listOfCordinates)
```

### WL information (Only run <span style='color:red'> before </span> mono. MMI imaging, no power info)

```python
fields = ['monoIndex','NKTIndex','monoOutpoint','mono_peakX', 'NKTpeakX', 'NKTSetpoint']
filename = "matching WL points list.csv"

with open(filename, 'w',newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames = fields) 
    writer.writeheader() 

for cord in listOfCordinates:    
    with open(filename, 'a',newline='') as csvfile: 
        # creating a csv dict writer object 
        writer = csv.writer(csvfile) 
        writer.writerow([cord[0],cord[1],mono_outpoints['outpoint'][cord[0]],mono_list['peakX'][cord[0]],\
                         NKT_list['peakX'][cord[1]],NKT_list['setpoint'][cord[1]]]) 
```

### Power information (Only run <span style='color:red'> after </span> mono. MMI imaging, power info)

```python
fields = ['monoIndex','NKTIndex','monoOetpoint','mono_peakX', 'NKTpeakX', 'NKTSetpoint','monoPowerLastIter','monoPowerAvgIter']
filename = "matching WL points list.csv"

with open(filename, 'w',newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames = fields) 
    writer.writeheader() 
    
for cord in listOfCordinates:    
#     Lastpower=MMI_power[(MMI_power['iter']==itera-1) & (MMI_power['WL']==mono_outpoints["outpoint"][cord[0]])]['power'].item()
    Lastpower=MMI_power[(MMI_power['iter']==itera-1) & (MMI_power['WL']==mono_outpoints["outpoint3"][cord[0]])]['power'].item()
    Avgpower=np.mean(MMI_power[MMI_power['WL']==mono_outpoints["outpoint3"][cord[0]]]['power'])
###     print(np.mean(MMI_power[MMI_power['WL']==mono_outpoints["outpoint"][cord[0]]]['power']))
    
    with open(filename, 'a',newline='') as csvfile: 
        # creating a csv dict writer object 
        writer = csv.writer(csvfile) 
        writer.writerow([cord[0],cord[1],mono_outpoints['outpoint3'][cord[0]],mono_list['peakX'][cord[0]],\
                         NKT_list['peakX'][cord[1]],NKT_list['setpoint'][cord[1]],\
                        Lastpower, Avgpower]) 
```
